<p><b> Checking if outliers are there </p> </b>

```python
#describing outliers present in our dataset
def outliers(x):
    q1=np.percentile(x,25)
    q3=np.percentile(x,75)
    iqr=q3-q1
    lower=q1-(1.5*iqr)
    upper=q3+(1.5*iqr)
    outlier_indics=list(x.index[(x<lower)|(x>upper)])
    outlier_value=list(x[outlier_indics])
    
    return outlier_indics,outlier_value
indics,value=outliers(data.TotalCharges)
print(indics)
print(value)
```

<p><b>Spliting our data into train andd test</b></p>

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, Y,random_state = 2,test_size=0.3)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

<p><b>Using RandomForest classifier </b></p>

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier( min_samples_leaf = 2,
                            min_impurity_decrease = 0.005,
                           verbose = 10, class_weight='balanced')
# build grid search parameters
parms = { 'min_samples_split'  : [2,5,10],
          'max_features' : [0.8,0.7,"auto"],
          'n_estimators' : [50,100,150,200],
          'bootstrap':[True, False]} 
# perform grid search
from sklearn.model_selection import GridSearchCV
CV = GridSearchCV(estimator = RF,param_grid = parms, scoring = 'roc_auc',cv=5, refit=True,
                           return_train_score = True, verbose=10,n_jobs=1)
CV.fit(train_x,train_y)
```

```python
print(CV.best_estimator_)
```

<b> This may be some important features for the would be churner <b> 

```python
feature_importances = pd.DataFrame(CV.best_estimator_.feature_importances_,
                                   index = train_x.columns,
                                    columns=['importance']).sort_values('importance',ascending=False)
print(feature_importances)
```

```python
# Preidct on Test data and check the confusion matrix
predictTest = CV.predict(test_x)
print(predictTest[0:10])
```

```python
predictProb = CV.predict_proba(test_x)
print(predictProb[0:10])
PProb = pd.DataFrame(predictProb)
```

```python
# generate confusion matrix
# import confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report
print (confusion_matrix(test_y,predictTest))
print (classification_report(test_y,predictTest))
```

```python
# generate ROC curve
from sklearn.metrics import roc_auc_score, roc_curve
# AUC for class 1
AUC = roc_auc_score(test_y, PProb[1])
print(AUC)
# AUC for class 0PProb[1]
#AUC0 = roc_auc_score(Y_test, PProb[0])
#print(AUC0)
# define variables for True Positive Rate and Falst Positive Rate & threshold value
TPR = dict()
FPR = dict()
THR = dict()
FPR, TPR, THR = roc_curve(test_y, PProb[1])
```
